Gerekli Kütüphanelerin eklenmesi

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Veri Setinin oluşturulması

```python
ilan_fiyat=np.array([20,25,30,35,40,45,50,55,60,65])
satis_miktari=np.array([40,45,55,60,65,75,80,85,95,100])
```

ilan_fiyat Göster

```python
ilan_fiyat
```

satis_miktari göster

```python
satis_miktari
```

```python
x=ilan_fiyat.reshape(-1,1)
```

```python
x
```

```python
y=satis_miktari
```

```python
y
```

```python
model=LinearRegression()
model.fit(x,y)
```

Verileri Grafiğe Dönüştürme

```python
plt.scatter(ilan_fiyat, satis_miktari,label='VeriSeti')
plt.show()
```

Yapay Zeka Tahmini Yaptırma

```python
ilan_fiyat_tahmin=np.linspace(20,70,100).reshape(-1,1)
```

```python
satis_miktari_tahmin=model.predict(ilan_fiyat_tahmin)
```

```python
satis_miktari_tahmin
```

```python
plt.plot(ilan_fiyat_tahmin, satis_miktari_tahmin)     #, color='red',label='RegresyonDogrusu')
plt.scatter(ilan_fiyat, satis_miktari,label='VeriSeti')
plt.legend()
plt.show()
```
