Gerekli kütüphaneleri projemize ekliyoruz

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Rastgele veriseti oluşturma

### Ev fiyatlarına göre metrekare hesabı

```python
np.random.seed(0)
n=100
metrekare = np.round(50 + 50 * np.random.rand(n,1))
fiyat = np.round (100000 + 5000 * metrekare + np.random.rand(n,1) * 25000)
```

Metrekare - fiyat **dağılım grafiği**

```python
plt.scatter(metrekare,fiyat)
```

# MODEL oluşturma

```python
model = LinearRegression()
model.fit(metrekare , fiyat)
```

Tahmin Çizgisi Oluşturma

```python
fiyat_tahmini = model.predict(metrekare)
```

Veri Noktalarını ve tahmin çizgisini(Dogrusal regresyon çizgisini) Görselleştirme

```python
plt.scatter(metrekare,fiyat,color='blue',label='Veri Noktaları')
plt.plot(metrekare,fiyat_tahmini,color='red',label='Dogrusal Regresyon Cizgisi')
plt.xlabel('Metrekare')
plt.ylabel('Fiyat')
plt.title('Ev fiyatlarına göre metrekare dağılımı ve regresyon çizgisi')
plt.legend()
plt.show()
```
