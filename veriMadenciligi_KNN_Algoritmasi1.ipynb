### KNN algoritması ile iris sınıflama örneği

https://archive.ics.uci.edu/dataset/53/iris

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
```

IRIS verisetinin projeye indirilmesi. Burada veriseti yüklemesi manuel de yapılabildiği gibi aşağıdaki şekilde hazır olarak da indirilebilir.

```python
from sklearn.datasets import load_iris
#iris adında bir değişken oluşturuyorum
iris = load_iris()
#özellikler (yaprak boyları vs)
x=iris.data
#çiçek türü (virginica, setona, versicolor)
y=iris.target
```

```python
x
```

```python
y
```

Verisetini %30 test %70 eğitim olmak üzere rastgele parçalardan bölüyoruz.

```python
#Eğitim ve Test veriseti diye bölme işlemi yapılıyor
xEgitim, xTest, yEgitim, yTest = train_test_split(x,y,test_size=0.3, random_state=1)

#Özellikleri ölçeklendir
olceklendir=StandardScaler()
xEgitim=olceklendir.fit_transform(xEgitim)
xTest=olceklendir.transform(xTest)
```

## KNN Modelinin Eğitimi

```python
k=3
knn=KNeighborsClassifier(n_neighbors=k)

# modelin eğitilmesi
knn.fit(xEgitim,yEgitim)
```

```python
# test verisi ile tahmin yapma
yTahmin = knn.predict(xTest)
```

```python
yTest
```

```python
yTahmin
```

%30 test için ayrılan verilerin gerçek değerlerini zaten biliyorduk.

Ama Yapay zekaya bu %30 u tahmin ettirdik ve yukarıdaki sonucu üretti.

Şimdi tahmin ettiği sonuçlar ile gerçek bildigimiz sonuçları kıyaslayalım.

```python
# Dogruluk Skoru

accuracy = accuracy_score(yTest,yTahmin)
print("Model Dogruluk Skoru:",accuracy)
```

```python
print("Sınıflandırma Raporu")
print(classification_report(yTest,yTahmin))
```
