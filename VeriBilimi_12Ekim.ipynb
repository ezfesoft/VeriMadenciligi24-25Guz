```python
# Sezar Tekniği
acik_mesaj='baba dede aba'
harfler=['a','b','c','ç','d','e','f','g','ğ','h']
sezar_sayisi=int(input('Sezar:'))
sifreli_mesaj=''

for harf in acik_mesaj:
  if(harf==' '):
    sifreli_mesaj +=' '
  index=0
  for dizi in harfler:
    if(harf == dizi):
      sifreli_mesaj += harfler[(index+sezar_sayisi)]
    index+=1

print('------------')
print(sifreli_mesaj)
```
