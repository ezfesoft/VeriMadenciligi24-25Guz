# Pandas ile Eksik Veri İşleme

**isnull()** -> Eksik veriler için True dönütü veriri

**notnull()** -> eksik veriler için False dönütü verir

**dropna()** -> Eksik verileri filtreliyip dönüt verir

**fillna()** -> Eksik verileri doldurup dönüt verir


```python
data=pd.Series([1, np.nan,"merhaba",None])
```

```python
data.isnull()
```

```python
data.notnull()
```

```python
data[data.notnull()]
```

```python
data.dropna()
```

```python
data
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
dataFrame2=pd.DataFrame([[1,np.nan,2],[2,3,5],[np.nan,4,6]])
```

```python
dataFrame2
```

```python
dataFrame2.dropna()
```

```python
dataFrame2.dropna(axis='columns')
```

```python
dataFrame2
```

```python
dataFrame2[3]=np.nan
```

```python
dataFrame2
```

```python
dataFrame2.dropna(axis='columns',how='all')
```

```python
dataFrame2
```

```python
dataFrame2.dropna(axis='rows',thresh=3)
```

# Nan Verilerin Doldurulması

```python
data3=pd.Series([1,np.nan,2,None,3],
                index=['a','b','c','d','e'])
```

```python
data3
```

```python
data3.fillna(0)
```

```python
data3
```

```python
data3.fillna(method='ffill')
```

```python
data3
```

```python
data3.fillna(method='ffill',inplace=True)
```

```python
data3
```
